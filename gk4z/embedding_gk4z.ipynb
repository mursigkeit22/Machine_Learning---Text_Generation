```python
import gensim, logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
import pandas as pd
from gensim.test.utils import common_texts, get_tmpfile
```

```python
df = pd.read_csv('gk4z_rus_clean.csv')
```

```python
df
```

```python
lines = []
for line in df.clean_text:
    lines.append(line.split())
```

```python
len(lines)
```

```python
path = get_tmpfile("gk4z_ruc_clean.model")
```

```python
model = gensim.models.Word2Vec(lines, min_count=2, workers=10, size=200)
```

```python
model.save("gk4z_ruc_clean.model")
```
